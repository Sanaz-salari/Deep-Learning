```python


# Make a prediction with weights
# To check y for each row to know whether w0+w1*x1+w2*x2 for each row is greater than 0, so y for that row would be 1, otherwise 0.
def predict(row, weights):
	activation = weights[0] # This line shows the linear summation value in a neuron. The activation function has not been performed on it. When it enters the activation function, it returns 1 or 0.
	for i in range(len(row)-1):   
		activation += weights[i + 1] * row[i]
	return 1.0 if activation >= 0.0 else 0.0  # here it is step function activation and it returns y.

  
# Estimate Perceptron weights using stochastic gradient descent

# train: training dataset ---> [x11, x12, y1; x21, x22, y2; ... ; xn1, xn2, yn], a matrix with n*3 dimensions
# l_rate: learning rate ---> Used to limit the amount each weight is corrected each time it is updated.
# n_epoch:  The number of times to run through the training data while updating the weight.
def train_weights(train, l_rate, n_epoch):
	weights = [0.0 for i in range(len(train[0]))] # [0.0, 0.0, 0.0]
	for epoch in range(n_epoch):
		sum_error = 0.0

# To update bias each time a row is entered, although bias is not associated with a specific input data
		for row in train:
			prediction = predict(row, weights)
			error = row[-1] - prediction
			sum_error += error**2
			weights[0] = weights[0] + l_rate * error # bias is updated and not associated with a specific input data

# To update other weights
			for i in range(len(row)-1): # len(row)-1 is 2, because we need to update weights only for x1 and x2. i=0,1
				weights[i + 1] = weights[i + 1] + l_rate * error * row[i] # row[i] is : dataset[row][i] : i=0 is x1 and i=1 is x2
        # in the above line since it doesn't calculate w0, so each w(i+1) is calculated based on its initialized w(i+1) which was 0 
		print('>epoch=%d, lrate=%.3f, error=%.3f' % (epoch, l_rate, sum_error))
	return weights

dataset = [[2.7810836,2.550537003,3.5, 0],
	[1.465489372,2.362125076,2.5, 0],
		[3.396561688,4.400293529,0],
	[0,3,0]]
l_rate = 0.1 # learning rate is a positive constant less than 1
n_epoch = 5
weights = train_weights(dataset, l_rate, n_epoch)


# Perceptron Algorithm With Stochastic Gradient Descent
def perceptron(train, test, l_rate, n_epoch):
	predictions = list() #To generate an empty list
	weights = train_weights(train, l_rate, n_epoch) # first, recall train-weights function to update weights
	for row in test:
		prediction = predict(row, weights) # second, generate predictions
		predictions.append(prediction)
	return(predictions)

```

```python
#The second way of modelling a perceptron
def perceptron2(X, y, lr, epochs):
    
    # X --> Inputs.
    # y --> labels/target.
    # lr --> learning rate.
    # epochs --> Number of iterations.
    
    # m-> number of training examples
    # n-> number of features 
    m, n = X.shape
    
    # Initializing parapeters(theta) to zeros.
    # +1 in n+1 for the bias term.
    theta = np.zeros((n+1,1))
    
    # Empty list to store how many examples were 
    # misclassified at every iteration.
    n_miss_list = []
    
    # Training.
    for epoch in range(epochs):
        
        # variable to store #misclassified.
        n_miss = 0
        
        # looping for every example.
        for idx, x_i in enumerate(X):
            
            # Insering 1 for bias, X0 = 1.
            x_i = np.insert(x_i, 0, 1).reshape(-1,1)
            
            # Calculating prediction/hypothesis.
            y_hat = step_func(np.dot(x_i.T, theta))
            
            # Updating if the example is misclassified.
            if (np.squeeze(y_hat) - y[idx]) != 0:
                theta += lr*((y[idx] - y_hat)*x_i)
                
                # Incrementing by 1.
                n_miss += 1
        
        # Appending number of misclassified examples
        # at every iteration.
        n_miss_list.append(n_miss)
        
    return theta, n_miss_list
```
